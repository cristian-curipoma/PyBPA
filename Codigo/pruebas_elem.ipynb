```python
sec_est_agua = Get_dry_data('./Datos/Secado/secado_estufa_aguacate.csv', ';')
sec_est_hojas_agua = Get_dry_data('./Datos/Secado/secado_hojas_estufa_aguacate.csv', ';')
```

```python
df_dry_est_ma = sec_est_agua.Get_dry_df(repeats=3, surfaces=['Metal'], lengths=['10 cm', '20 cm', '30 cm'])
df_dry__est_ha = sec_est_hojas_agua.Get_dry_df(repeats=2, surfaces=['Metal'], lengths=['Unique'])
```
